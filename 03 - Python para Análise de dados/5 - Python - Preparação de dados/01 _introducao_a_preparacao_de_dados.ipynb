### Análise Exploratorio de Dados (AED) 

```python
# importando bibliotecas
import pandas as pd
```

```python
# Lendo o arquivo CSV e armazenando em um DataFrame
df = pd.read_csv(r'C:\Users\rodri\dev\ciencia_de_dados_v2\assets\clientes-v2.csv')

# Exibindo os 10 primeiro registro do DataFrame
df.head(10)
```

```python
# Exibindo os 10 ultimos registros do DataFrame
df.tail(10)
```

```python
# Ajustando o formado da data
# pode ser feito na area de tratamento de dados 
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y', errors='coerce')
```

```python
# Exibindo informações do DataFrame
df.info()
```

```python
# Exibindo os dados nulors do DataFrame
print('Analise dos dados nulos')
df.isnull().sum()
```

```python
# Exibindo os dados nulos do DataFrame em porcentagem 
print('Analise dos dados nulos em porcentagem')
df.isnull().mean() * 100
```

```python
# removendo dados nulos
df.dropna(inplace=True)

# Confirmando a remocao dos dados nulos
df.isnull().sum() 
```

```python
# Analisando dados duplicados
df.duplicated().sum()
```

```python
# Analisando dados únicos
df.nunique()
```

```python
# Estatística de dados
df.describe()
```

```python
# Montando dataframe com dados utilizados
df = df[['idade', 'data', 'estado', 'salario', 'nivel_educacao', 'numero_filhos', 'estado_civil', 'area_atuacao']]
df.head()
```

```python
# Salvando o DataFrame em um arquivo CSV
df.to_csv(r'C:\Users\rodri\dev\ciencia_de_dados_v2\assets\clientes-v2-tratado.csv', index=False)
```
