# Exemplos de Gráficos

Na aula anterior, você compreendeu o uso da biblioteca Seaborn. Nesta aula, você vai:
- Aplicar técnicas de visualização de dados para realizar o agrupamento de dados e analisar a contagem de diferentes categorias;

- Criar gráficos visuais complexos utilizando as bibliotecas Matplotlib e Seaborn.

```python
# Importando bibliotecas e lendo arquivo CSV
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('clientes-v3-preparado.csv')
df.head()
```

```python
df_corr = df[['salario', 'idade', 'anos_experiencia', 'numero_filhos', 'nivel_educacao_cod', 'area_atuacao_cod', 'estado_cod']].corr()
```

```python
# Heatmap de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr.corr(), annot=True, fmt=".2f")
plt.title('Mapa de Calor da Correlação entre Variáveis')
plt.show()
```

```python
# Countplot
sns.countplot(x='estado_civil', data=df)
plt.title('Distribuição do Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Contagem')
plt.show()
```

```python
# Countplot com legenda
sns.countplot(x='estado_civil', hue='nivel_educacao', data=df)
plt.title('Distribuição do Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Contagem')
plt.legend(title='Nível de Educação')
plt.show()
```
