```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_preparados.csv')
df
```

```python
# verificando se os dados est√£o corretos
df.info()
```

```python
#verificando dados nulos em porcentagem
df.isnull().sum() / df.shape[0] * 100
```

```python
# eliminando linhas que contem dados nulos
df.dropna(inplace=True)
df
```

```python
# verificando dados duplicados
df.duplicated().sum()
```

```python
# Gr√°fico de dispers√£o - pre√ßo unit√°rio vs quantidade vendida
plt.figure(figsize=(10, 6))
plt.scatter(df['Pre√ßo'], df['Qtd_Vendidos'], alpha=0.5)
plt.title('Gr√°fico de Dispers√£o: pre√ßo unit√°rio vs quantidade vendida')
plt.xlabel('Pre√ßo Unit√°rio')
plt.ylabel('Quantidade Vendida')
plt.grid()
plt.show()
```

```python
# Mapa de calor
# Criando uma matriz de correla√ß√£o
plt.Figure(figsize=(10, 5))
correlacao = df[['Pre√ßo', 'Nota']].corr()
plt.title('Correla√ß√£o - pre√ßo vs nota')
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.tight_layout()
```

```python
# Gr√°fico de barras - quantidade vendida por temporada
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Temporada', order=df['Temporada'].value_counts().index, palette='viridis')
plt.title('Quantidade Vendida por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Gr√°fico de pizza - distribui√ß√£o de produtos por quantidade de vendas
plt.figure(figsize=(10, 6))
df['Qtd_Vendidos'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Distribui√ß√£o de Produtos por Qtd_Vendidos')
plt.ylabel('')
plt.axis('equal')  # Para deixar o gr√°fico em formato de c√≠rculo
plt.show()
```

```python
# Gr√°fico de densidade - pre√ßo unit√°rio
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Pre√ßo'], fill=True, color='blue', alpha=0.5)
plt.title('Distribui√ß√£o de Pre√ßo Unit√°rio')
plt.xlabel('Pre√ßo Unit√°rio')
plt.ylabel('Densidade')
plt.grid()
plt.show()
```
