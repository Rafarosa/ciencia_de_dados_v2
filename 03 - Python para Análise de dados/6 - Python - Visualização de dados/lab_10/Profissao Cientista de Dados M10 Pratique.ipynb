```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_preparados.csv')
df
```

```python
# verificando se os dados estão corretos
df.info()
```

```python
#verificando dados nulos em porcentagem
df.isnull().sum() / df.shape[0] * 100
```

```python
# eliminando linhas que contem dados nulos
df.dropna(inplace=True)
df
```

```python
# verificando dados duplicados
df.duplicated().sum()
```

```python
# Gráfico de dispersão - preço unitário vs quantidade vendida
plt.figure(figsize=(10, 6))
plt.scatter(df['Preço'], df['Qtd_Vendidos'], alpha=0.5)
plt.title('Gráfico de Dispersão: preço unitário vs quantidade vendida')
plt.xlabel('Preço Unitário')
plt.ylabel('Quantidade Vendida')
plt.grid()
plt.show()
```

```python
# Mapa de calor
# Criando uma matriz de correlação
plt.Figure(figsize=(10, 5))
correlacao = df[['Preço', 'Nota']].corr()
plt.title('Correlação - preço vs nota')
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.tight_layout()
```

```python
# Gráfico de barras - quantidade vendida por temporada
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Temporada', order=df['Temporada'].value_counts().index, palette='viridis')
plt.title('Quantidade Vendida por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Gráfico de pizza - distribuição de produtos por quantidade de vendas
plt.figure(figsize=(10, 6))
df['Qtd_Vendidos'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Distribuição de Produtos por Qtd_Vendidos')
plt.ylabel('')
plt.axis('equal')  # Para deixar o gráfico em formato de círculo
plt.show()
```

```python
# Gráfico de densidade - preço unitário
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Preço'], fill=True, color='blue', alpha=0.5)
plt.title('Distribuição de Preço Unitário')
plt.xlabel('Preço Unitário')
plt.ylabel('Densidade')
plt.grid()
plt.show()
```
