```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# lendo arquvivo csv
df = pd.read_csv('clientes-v3-preparado.csv')
df.head(10)
```

Histograma

Um histograma é uma espécie de gráfico de barras que demonstra uma distribuição de frequências. No histograma, a base de cada uma das barras representa uma classe e a altura representa a quantidade ou frequência absoluta com que o valor de cada classe ocorre.

```python
# histograma de salário
plt.hist(df['salario'])
```

Histograma com Parâmetros

```python
# histograma com parametros
plt.figure(figsize=(10, 5))
plt.hist(df['salario'], bins=100, color='green', alpha=0.8)
plt.title('Histograma - Distribuição de Salário')
plt.xlabel('Salário')
plt.xticks(ticks=range(0, int(df['salario'].max())+2000,2000))
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

Multiplos gráficos

```python
# Multiplos Graficos
# Criando multiplos graficos
plt.Figure(figsize=(10, 5))

# Criando o primeiro gráfico
plt.subplot(2, 2, 1) # 2 linhas, 2 colunas, 1º gráfico

# Grafico de dispersão
plt.scatter(df['salario'], df['salario'])
plt.title('Dispersão - Salário vs Salário')
plt.xlabel('Salário')
plt.ylabel('Salário')

plt.subplot(1, 2, 2) # 1 linhas, 2 colunas, 2º gráfico
plt.scatter(df['salario'], df['anos_experiencia'], color='#5883a8', alpha=0.6, s=30)
plt.title('Dispersão - Salário vs Anos de Experiência')
plt.xlabel('Salário')
plt.ylabel('Anos de Experiência')

# Mapa de calor

corr = df[['salario', 'anos_experiencia']].corr()
plt.subplot(2, 2, 3) # 1 linhas, 2 colunas, 1º gráfico
plt.title('Correlação - Salário vs Anos de Experiência')
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.tight_layout() # Ajusta o layout para evitar sobreposição de gráficos
```
